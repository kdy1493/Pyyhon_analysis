## Numpy(넘파이) 개요
- Numpy는 Numerical Python의 줄임말
- 파이썬에서 산술 계산을 위한 가장 중요한 필수 패키지
- 효율적인 다차원 배열 ndarray는 빠른 배열 계산 지원
- 유연한 브로드캐스팅 기능 제공
- 반복문을 작성할 필요 없이 전체 데이터 배열을 빠르게 계산
