### Workshop (거리 계산 - numpy 사용)

**거리계산**
- 리스트를 이용하여 점 사이의 거리를 계산 해 보자.
- 직교 좌표 위에서 A는 (1, 1), B는 (3, 2), C는 (5,7)일 때 X(2, 3)와  A/B/C와의 거리를 각각 구하여라.
- 위 코드블럭을 이용하여 distMeasure 함수를 정의하라.

**파이썬 이용한 버전**

```python
# 점 사이 거리 구하는 함수 
def disMeasure (a , b) :
    result = ((b[0] - a[0])**2 + (b[1]-a[1])**2)**0.5 
    return result
arr1 = [(1,1),(3,2) ,(5,7)]

# a = [1,1]
# b = [3,2] 
# c = [5,7] 
x= (2,3) 


for arr in arr1 : 
    print(disMeasure(arr, x))
    
```

**넘파이 이용한 버전**

```python
arr2 = np.array(arr1)
x_2 = np.array(x)

np.sqrt(np.sum((arr2 - x_2) **2,axis =1))
```
